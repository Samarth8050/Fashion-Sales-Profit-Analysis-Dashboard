```python
import pandas as pd
import numpy as np

df=pd.read_csv("C:/Users/PCSS/OneDrive/Desktop/PROJECT/SNITCH/Snitch_Fashion_Sales_Uncleaned.csv")
df.head(10)
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# fill missing numeric values
df['Units_Sold'] = df['Units_Sold'].fillna(1)
```

```python
# fill missing discount
df['Discount_%'] = df['Discount_%'].fillna(0)
```

```python
# fix city names
df['City'] = df['City'].str.title()
```

```python
# convert date
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

```python
# calculate sales
df['Sales_Amount'] = df['Units_Sold'] * df['Unit_Price'] * (1 - df['Discount_%'])
```

```python
df['Unit_Price'] = df['Unit_Price'].fillna(df['Unit_Price'].median())
print(df.isnull().sum())
```

```python
df['Segment'] = df['Segment'].fillna("Unknown")
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Order_Date'] = df['Order_Date'].fillna(method='ffill')
```

```python
df['Sales_Amount'] = df['Units_Sold'] * df['Unit_Price'] * (1 - df['Discount_%'])
```

```python
df.to_csv("Cleaned_Fashion_Sales.csv", index=False)
```
